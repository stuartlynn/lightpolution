```python
from astropy.stats import sigma_clipped_stats
from photutils.datasets import load_star_image
from astropy.visualization.mpl_normalize import ImageNormalize
from astropy.visualization import SqrtStretch
from rasterio.windows import Window, from_bounds
from rasterio.plot import show, show_hist
import numpy as np
import matplotlib.pyplot as plt
import rasterio
```

```python
image = rasterio.open('../images/SVDNB_npp_20210301-20210331_global_vcmslcfg_v10_c202104061200.avg_rade9h.tif')
```

```python
image.shape
```

```python
image.bounds
```

```python
us_bounding = (-126.0,22.9,-50.98,54.59)
```

```python
us_window = from_bounds(us_bounding[0], us_bounding[1], us_bounding[2], us_bounding[3], image.transform)
```

```python
data = image.read(1, window= us_window)
```

```python
data[data<0] = 0
```

```python
show(np.log(data),with_bounds=True)
```

```python
mean, median, std = sigma_clipped_stats(data, sigma=3.0)  
```

```python
print(mean,median,std)
```

```python
from photutils.detection import DAOStarFinder
```

```python
daofind = DAOStarFinder(fwhm=3.0, threshold=7.*std) 
```

```python
sources = daofind(data.T-median)
```

```python
psources = sources.to_pandas()
```

```python
us_window
```

```python
data.T.shape
```

```python
psources = psources.assign(x = (psources.xcentroid*us_window.width)/data.shape[0] + us_window.row_off, 
                           y = (psources.ycentroid*us_window.height)/data.shape[1] + us_window.col_off)
```

```python
def to_coord(x,y):
    return image.xy(x,y)
```

```python
lngs = psources[['xcentroid','ycentroid']].apply(lambda x: to_coord(x.xcentroid,x.ycentroid)[1] ,axis=1)
lats = psources[['xcentroid','ycentroid']].apply(lambda x: to_coord(x.xcentroid,x.ycentroid)[0] ,axis=1)
```

```python
psources = psources.assign(lat=lats, lng=lngs)
```

```python
plt.scatter(psources.lng, psources.lat, s=0.01)
```

```python
psources.to_csv("sources.csv",index=False)
```
